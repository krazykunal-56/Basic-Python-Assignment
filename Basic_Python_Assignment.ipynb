```python
# Basic Python Assignment: Arithmetic Operations and Type Conversions

# 1. Write a Python program to add two numbers entered by the user
def add_numbers():
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    print("The sum is:", num1 + num2)

# 2. Write a Python program to subtract one number from another entered by the user
def subtract_numbers():
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    print("The difference is:", num1 - num2)

# 3. Write a Python program to multiply two numbers entered by the user
def multiply_numbers():
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    print("The product is:", num1 * num2)

# 4. Write a Python program to divide one number by another entered by the user
def divide_numbers():
    num1 = float(input("Enter numerator: "))
    num2 = float(input("Enter denominator: "))
    if num2 != 0:
        print("The result of division is:", num1 / num2)
    else:
        print("Division by zero is not allowed")

# 5. Write a Python program to calculate the remainder when one number is divided by another
def remainder_numbers():
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))
    print("The remainder is:", num1 % num2)

# 6. Write a Python program to calculate the power of a number (x^y)
def power_numbers():
    x = float(input("Enter base number: "))
    y = float(input("Enter exponent: "))
    print(f"{x} raised to the power of {y} is:", x ** y)

# 7. Write a Python program that converts an integer into a float
def int_to_float():
    num = int(input("Enter an integer: "))
    print("Converted to float:", float(num))

# 8. Write a Python program that converts a float into an integer
def float_to_int():
    num = float(input("Enter a float number: "))
    print("Converted to integer:", int(num))

# 9. Write a Python program that converts a number into a string
def number_to_string():
    num = input("Enter a number: ")
    print("Converted to string:", str(num))

# 10. Write a Python program that converts a string into an integer
def string_to_int():
    num_str = input("Enter a string representing an integer: ")
    print("Converted to integer:", int(num_str))

# 11. Write a Python program that converts a string into a float
def string_to_float():
    num_str = input("Enter a string representing a float: ")
    print("Converted to float:", float(num_str))

# 12. Write a Python program to take user input as a string and display its type after converting it to integer, float, and string
def input_type_conversion():
    user_input = input("Enter something: ")
    try:
        as_int = int(user_input)
    except:
        as_int = "Conversion not possible"
    try:
        as_float = float(user_input)
    except:
        as_float = "Conversion not possible"
    as_str = str(user_input)

    print("Original input type:", type(user_input))
    print("After converting to integer:", type(as_int) if isinstance(as_int,int) else as_int)
    print("After converting to float:", type(as_float) if isinstance(as_float,float) else as_float)
    print("After converting to string:", type(as_str))

# Main menu
while True:
    print("\n--- Basic Python Operations ---")
    print("1. Add two numbers")
    print("2. Subtract two numbers")
    print("3. Multiply two numbers")
    print("4. Divide two numbers")
    print("5. Remainder of division")
    print("6. Power of a number (x^y)")
    print("7. Convert integer to float")
    print("8. Convert float to integer")
    print("9. Convert number to string")
    print("10. Convert string to integer")
    print("11. Convert string to float")
    print("12. Input type conversion demonstration")
    print("13. Exit")

    choice = input("Enter your choice (1-13): ")

    if choice == "1":
        add_numbers()
    elif choice == "2":
        subtract_numbers()
    elif choice == "3":
        multiply_numbers()
    elif choice == "4":
        divide_numbers()
    elif choice == "5":
        remainder_numbers()
    elif choice == "6":
        power_numbers()
    elif choice == "7":
        int_to_float()
    elif choice == "8":
        float_to_int()
    elif choice == "9":
        number_to_string()
    elif choice == "10":
        string_to_int()
    elif choice == "11":
        string_to_float()
    elif choice == "12":
        input_type_conversion()
    elif choice == "13":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 13.")
```
